### Tree General Event Data

```python
# Event data from tree general
path_tree_general = os.getcwd() + "/data" + "/Tree General"
filenames_tree_general = glob.glob(path_tree_general + "/*EVD.*")

dfs_tree_general = []
for filename in filenames_tree_general:
    dfs_tree_general.append(pd.read_csv(filename, sep = '\t', header = None, error_bad_lines=False, warn_bad_lines=False))
```

```python
tree_general_evd_df = pd.concat(dfs_tree_general, ignore_index=True)
```

```python
tree_general_evd_df.shape
```

### Calculating Sacade Length in Tree General

```python
tree_general_dx = (tree_general_fxd_df['screen_x'] - tree_general_fxd_df['screen_x'].shift())
tree_general_dy = (tree_general_fxd_df['screen_y'] - tree_general_fxd_df['screen_y'].shift())
tree_general_fxd_df['saccade_length'] = np.sqrt(tree_general_dx**2 + tree_general_dy**2)
```

```python
tree_general_fxd_df.head(10)
```
